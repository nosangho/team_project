
리스트에서 상품 코드를 획득하여, 해당 상품 페이지의 URL을 생성한다. 


상품리스트 URL 형식

    
page번호 바꿔주면 해당 페이지로 이동한다. 
basic 
http://www.wine21.com/13_WineSearch/winesearch01_list.php?sh_searchtype=1&sh_orderby=viewcnt|DESC&sh_MaxPage=100

add
&page={}&increment=0.format.page_index 로 해서 페이지 넘기도록 한다. 


소요시간 
(슬립 0.3)

* 1~17 :  25 
* 18 ~ 100 :  202
* 101 ~ 177 : 189.263000011

```python
import requests
import re
from bs4 import BeautifulSoup
from HTMLParser import HTMLParser
import pandas as pd
import numpy as np
import time

class GetUrl(object):
    def __init__(self, url):
        self.url = url
        
    # 상품 검색 리스트에서 크롤링 시작, 끝 페이지를 입력하여 부분적으로 리스트 저장 가능하도록 한다.     
    def pd_url(self, start_p, end_p):
        
        pd_list_url=[]
        
        for i in range(start_p , end_p+1):
            
            pd_list_url.append(self.url+'&page={}'.format(i))
            
        return pd_list_url
        
    
    def get_code(self, url_list):
        
        code = []
        for i in url_list:
            res = requests.get(i)
            soup = BeautifulSoup(res.text)
        
            for dd in soup.find_all('dd', attrs = {'class' : 'EngName'}):
                code.append(str(dd.a['href']))
            time.sleep(0.3)
            
        return code
    
    def make_url(self, url_list):

        pd_url = {}
        url_base = 'http://wine21.com/13_WineSearch/winesearch01_view.php?SelUno={}'
        
        for i in url_list:
            val = re.search(r'\d{5}', i).group()
            pd_url[val] = url_base.format(val)
        
        return pd_url
    
url ='http://wine21.com/13_WineSearch/winesearch01_list.php?sh_searchtype=1&sh_MaxPage=100'  


# 시간 체크용 코드 
# start_time = time.time()
link = GetUrl(url)

# 임시로 start, end 1,1 로 입력해둔다. 
pd_list = link.pd_url(1,1)

pd_code = link.get_code(pd_list)

pd_url = link.make_url(pd_code)

# 시간 체크용 코드 
# end_time = time.time()
# print end_time - start_time




    
```

```python
# 위에서 저장한 dict에서 튜플로 받아 datafrqme 형태로 변환하여 엑셀로 저장한다. 

data = pd_url.items()
df1= pd.DataFrame(data)


print df1

# 페이지를 나눠서 저장하는 경우에 파일 이름을 잊지 말고 변경해주세요. 
writer = pd.ExcelWriter('./list3.xlsx', engine=None)

df1.to_excel(writer, sheet_name='1', encoding ='utf-8')

writer.save()
```
